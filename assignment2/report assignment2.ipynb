#Introduction
Here is an analysis of Billboard Hot 100 Hits from 1958-present. /n 
The dataset which I based on contains all of the top 1 singles on Billboard Hot 100 chart and their basic information from 1958 to the present./n
**How This Chart Works:**/n
Radio Airplay + Sales Data + Streaming Data = HOT 100/n 
Every week's most popular songs across all genres, ranked by radio airplay audience impressions, as measured by Nielsen Music, sales data as compiled by Nielsen Music, and streaming activity data provided by online music sources. (cited from Billboard.com)

```python
import pandas as pd
from pyecharts import Bar
import csv
from pyecharts import Line
```

```python
df = pd.read_csv('Billboard Top1 1958-2018.csv')   # the dataset is a mixed data of my previous scaping from Billboard website and others work I download from GitHub. I processed the data and cleaned it. 
```

```python
df=df.rename(columns={'Unnamed: 0':'number'})
df=df.drop(axis = 1, labels=['number'])
df.head()
```


#Who are the pop stars over this 60 years?

We can see from the chart that **Mariah Carey** and **The Beatles** are the top 2 artists (band), and their performances far surpasses those of other singers, while other singers' performances are very close. /n However, from the perspective of the number of their top singles, the situation is very different.
```python
artists_counts=df['artist'].value_counts()[:10].sort_values(ascending=False) #sort values, these 10 artists occupy the position 1 for the most weeks
artists = pd.DataFrame(artists_counts)

M=df[df['artist']=='Mariah Carey']['song'].describe() 
B=df[df['artist']=='The Beatles']['song'].describe()
BOY=df[df['artist']=='Boyz II Men' ]['song'].describe()
Madonna=df[df['artist']=='Madonna']['song'].describe()
Whitney_Houston=df[df['artist']=='Whitney Houston']['song'].describe()
Michael_Jackson=df[df['artist']=='Michael Jackson']['song'].describe()
Drake=df[df['artist']=='Drake']['song'].describe()
The_Black_Eyed_Peas=df[df['artist']=='The Black Eyed Peas']['song'].describe()
Bee_Gees=df[df['artist']=='Bee Gees']['song'].describe()
Adele=df[df['artist']=='Adele']['song'].describe()
singer=[M, B, BOY, Madonna, Whitney_Houston, Michael_Jackson, Drake, The_Black_Eyed_Peas, Bee_Gees, Adele] 
list=[]  #From these 10 artists, I try to find the quantity of their top1 unique songs 
for i in singer:
    unique=i['unique']
    list.append(unique)
attr = artists.index
v1 = artists.artist
v2 = list
bar = Bar("Top 10 Aritists and Performance")
bar.add("Weeks on Top 1", attr, v1, mark_line=["average"], is_label_show=True)
bar.add("Quantity of Top1 Songs", attr, v2, mark_point=["min", "max"], is_label_show=True)
bar
```

We can see that Mariah Carey has been active on the list for almost 18 years. The long career span is one of the reasons why she ranks the first in the dimension of weeks. /n However, **The Beatles** got this amazing performance in just six years and had the most number of top1 singles. #What about the variety of pop songs over time?
We can see that the number of unique top1 songs and that of the unique artists are almost the same, but in 1965 and 1989, the gaps between them are 10 and 9 relatively( which are large gaps). The possible reason is that there are some artists appear in the list for several times with different songs. /n 
Digging down, I find out there are some super stars in 1965 and 1989, such as **The Beatles**, in 1965, occupying the first position for 12 weeks with 5 unique songs, as well as **The Supremes**, in 1965, ranking the first for 6 weeks with 4 unique songs./n
**Furthermore, a surprising finding is that after 1991, the pop music is worse than the music of previous decades due to its lack of variety. We can observe that the numbers are under the average after 1991. In sum, the Billboard Hot 100 Chart is becoming less diverse over time in terms of unique artists and songs.**

```python
years=['1959', '1960', '1961', '1962', '1963','1964','1965','1966','1967','1968','1969','1970','1971','1972','1973','1974','1975','1976','1977','1978','1979','1980',
'1981','1982','1983','1984','1985','1986','1987','1988','1989','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003',
'2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018']
songs=[]
artists=[]
for i in years:
    uniques=df[df['date'].str.contains(i)]['song'].describe()
    uniquea=df[df['date'].str.contains(i)]['artist'].describe()
    unique_song=uniques['unique']
    unique_artist=uniquea['unique']
    songs.append(unique_song)
    artists.append(unique_artist)

L=[years,songs,artists]
zip3=zip(*L)

attr=years
v1 = songs
v2 = artists
line = Line("Unique change trend")
line.add("Unique Songs", attr, v1, mark_line=["average"], mark_point=['max','min'])
line.add('Unique Artists', attr, v2,mark_line=["average"],mark_point=['max','min'])
line
```
